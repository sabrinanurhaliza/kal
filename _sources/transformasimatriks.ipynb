# MATRIKS TRANSFORMATION

##### 1. Konsep Dasar Transformasi Matriks
Matrix transformation adalah cara mengubah vektor atau bentuk di bidang kartesian (koordinat 2D atau 3D) menggunakan perkalian matriks. Ini merupakan aplikasi penting dari aljabar linear dalam geometri, grafik komputer, fisika, dan teknik.

Jika sebuah matriks 𝐴 dikalikan dengan vektor 𝑥⃗, maka:

A𝑥⃗ = vektor baru hasil transformasi

##### 2. Perkalian Matriks dengan Vektor
Perkalian matriks dengan vektor menghasilkan vektor baru. Misalnya, jika:

A = 
$\begin{bmatrix}
 2 & -1 \\ 
 1 & 3 
 \end{bmatrix}$, 

A𝑥⃗ =
$\begin{bmatrix} 
 1 \\ -1
 \end{bmatrix}$

maka hasilnya:
$\begin{bmatrix}
 3 \\ -2
\end{bmatrix}$

Vektor hasil ini bisa diplot pada bidang kartesian untuk melihat efek visual transformasi.

##### 3. Interpretasi Geometris Transformasi

Setiap operasi matriks memiliki makna geometris tertentu:

Rotasi (perputaran): Mengubah arah vektor dengan sudut tertentu.

Refleksi (pencerminan): Mencerminkan vektor terhadap garis tertentu.

Stretch/Shrink (perenggangan/penyusutan): Mengubah panjang vektor dalam arah tertentu.

Shear (penggeseran): Mengubah bentuk objek dengan menggeser satu arah terhadap arah lainnya.

##### 4. Transformasi Terhadap Unit Square

Unit square adalah persegi satuan (ukuran 1x1) yang sering digunakan untuk menunjukkan bagaimana bentuk berubah karena transformasi. Jika vektor basis awal adalah:

$\vec{e}_1 =
\begin{bmatrix}
 1 \\
 0 
\end{bmatrix}$,
$\quad \vec{e}_2 = 
\begin{bmatrix}
 0 \\
 1 
\end{bmatrix}$

Maka hasil transformasi dari 𝐴 terhadap vektor ini akan membentuk bentuk baru (seperti jajaran genjang, persegi miring, dsb).
Contoh: Jika

$A = \begin{bmatrix}
 1 & 2 \\
 0 & 1
\end{bmatrix}$

maka unit square akan berubah menjadi bentuk yang ter-shear ke arah horizontal.

##### 5. Macam-Macam Matriks Transformasi Umum

Beberapa transformasi penting dan matriksnya:

Horizontal Stretch by $k$:
$\begin{bmatrix} k & 0 \\ 0 & 1 \end{bmatrix}$

Vertical Stretch by $k$:
$\begin{bmatrix} 1 & 0 \\ 0 & k \end{bmatrix}$

Horizontal Shear by $k$:
$\begin{bmatrix} 1 & k \\ 0 & 1 \end{bmatrix}$

Vertical Shear by $k$:
$\begin{bmatrix} 1 & 0 \\ k & 1 \end{bmatrix}$

Refleksi terhadap x-axis:
$\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$

Refleksi terhadap y-axis:
$\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}$

##### 6. Transformasi geometri dalam bidang 2 dimensi

Transformasi geometri dalam bidang 2 dimensi, seperti refleksi, rotasi, proyeksi, geseran (shear), dan peregangan (stretch), dapat secara efisien direpresentasikan dan dioperasikan menggunakan matriks berukuran 2x2. Ketika beberapa transformasi diterapkan secara berurutan pada suatu objek, transformasi gabungan dapat diperoleh dengan mengalikan matriks-matriks transformasi individual. Penting untuk memperhatikan urutan perkalian matriks karena operasi ini umumnya tidak bersifat komutatif, yang berarti urutan penerapan transformasi dapat mempengaruhi hasil akhir. Konsep transformasi matriks ini memiliki aplikasi yang luas dan mendasar dalam berbagai bidang teknologi dan ilmu pengetahuan, termasuk grafika komputer untuk memanipulasi objek visual, pemrosesan citra untuk analisis dan modifikasi gambar, serta robotika untuk perencanaan gerakan dan orientasi.

##### Soal 1
 Diberikan matriks $A = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix}$, vektor $\vec{x} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$, dan vektor $\vec{y} = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$. Kita perlu mensketsa $\vec{x}$, $\vec{y}$, $A\vec{x}$, dan $A\vec{y}$ pada bidang kartesius yang sama.

$x = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ : titik (1, 1)

$y = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$ : titik (-1, 2)

$Ax = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 1(1)+(-1)(1) \\ 2(1)+3 (1) \end{bmatrix} = \begin{bmatrix} 1  -1 \\ 2 + 3 \end{bmatrix} = \begin{bmatrix} 0 \\ 5 \end{bmatrix}$: Titik (0, 5)

$Ay = \begin{bmatrix} 1 & -1 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} -1 \\ 2 \end{bmatrix} = \begin{bmatrix} 1(-1)+(-1)(2) \\ 2(-1)+3(2) \end{bmatrix} = \begin{bmatrix} -1 & -2 \\ -2 &+& 6 \end{bmatrix} = \begin{bmatrix} -3 \\ 4 \end{bmatrix}$: Titik (-3, 4)

##### Soal 2
 Diberikan matriks $A = \begin{bmatrix} 2 & 0 \\ -1 & 3 \end{bmatrix}$, vektor $\vec{x} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$, dan vektor $\vec{y} = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$. Kita perlu mensketsa $\vec{x}$, $\vec{y}$, $A\vec{x}$, dan $A\vec{y}$ pada bidang kartesius yang sama.

$x = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ : titik (1, 1)

$y = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$ : titik (-1, 2)

$Ax = \begin{bmatrix} 2 & 0 \\ -1 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 2(1)&+&(0)(1) \\ -1(1)&+&3(1) \end{bmatrix} = \begin{bmatrix} 2 &+& 0 \\ -1 &+& 3 \end{bmatrix} = \begin{bmatrix} 2 \\ 2 \end{bmatrix}$: Titik (2, 2)

$Ay = \begin{bmatrix} 2 & 0 \\ -1 & 3 \end{bmatrix} \begin{bmatrix} -1 \\ 2 \end{bmatrix} = \begin{bmatrix} 2(-1)&+&(0)(2) \\ -1(-1)&+&3(2) \end{bmatrix} = \begin{bmatrix} -2 &+& 0 \\ 1 &+& 6 \end{bmatrix} = \begin{bmatrix} -2 \\ 7 \end{bmatrix}$: Titik (-2, 7)

##### Soal 5
Sketsa menunjukkan transformasi dari unit persegi. Kita perlu menemukan matriks A yang melakukan transformasi ini.
Perhatikan bagaimana vektor basis $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ (ujung kanan persegi) ditransformasikan menjadi vektor $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ (ujung kanan jajargenjang). Vektor basis $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$ (ujung atas persegi) ditransformasikan menjadi vektor $\begin{bmatrix} 1 \\ 3 \end{bmatrix}$ (ujung atas jajargenjang).

Matriks transformasi A memiliki kolom-kolom yang merupakan citra dari vektor-vektor basis. Oleh karena itu, matriks A adalah:

$A = \begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix}$

##### Soal 6
Sketsa menunjukkan transformasi dari unit persegi menjadi persegi lain yang diputar dan diperkecil.
vektor basis $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ (ujung kanan persegi) ditransformasikan menjadi vektor $\begin{bmatrix} 1 \\ -1 \end{bmatrix}$ (ujung kanan persegi yang ditransformasikan). Vektor basis $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$ (ujung atas persegi) ditransformasikan menjadi vektor $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ (ujung yang ditransformasikan).

Matriks transformasi A adalah:

$A = \begin{bmatrix} 1 & 1 \\ -1 & 1 \end{bmatrix}$
