# PERKALIAN SILANG DUA VEKTOR (CROSS PRODUCT)

Perkalian silang dua vektor atau cross product adalah operasi dalam aljabar vektor yang menghasilkan vektor baru yang tegak lurus (ortogonal) terhadap dua vektor asal dalam ruang tiga dimensi (3D).

Sifat-sifat Perkalian Silang
Hasil vektor: bukan skalar.

Tegak lurus terhadap kedua vektor asal.

Arah hasil ditentukan oleh aturan tangan kanan (right-hand rule).

Misalkan dua vektor:  
$$
\vec{v} = (v_1, v_2, v_3), \quad \vec{u} = (u_1, u_2, u_3)
$$

Maka:  
$$
\vec{v} \times \vec{u} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
v_1 & v_2 & v_3 \\
u_1 & u_2 & u_3
\end{vmatrix}
=
(v_2 u_3 - v_3 u_2)\mathbf{i} - (v_1 u_3 - v_3 u_1)\mathbf{j} + (v_1 u_2 - v_2 u_1)\mathbf{k}
$$

##### Contoh:

Jika  
$$
\vec{v} = (1, 0, 0), \quad \vec{u} = (0, 1, 0)
$$

Maka:  
$$
\vec{v} \times \vec{u} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 0 \\
0 & 1 & 0
\end{vmatrix}
=
(0 \cdot 0 - 0 \cdot 1)\mathbf{i} - (1 \cdot 0 - 0 \cdot 0)\mathbf{j} + (1 \cdot 1 - 0 \cdot 0)\mathbf{k}
= \mathbf{k}
$$

##### Identitas Cross Product:
$$
\begin{aligned}
\mathbf{i} \times \mathbf{j} &= \mathbf{k} \\
\mathbf{j} \times \mathbf{k} &= \mathbf{i} \\
\mathbf{k} \times \mathbf{i} &= \mathbf{j} \\
\\
\mathbf{j} \times \mathbf{i} &= -\mathbf{k} \\
\mathbf{i} \times \mathbf{k} &= -\mathbf{j} \\
\mathbf{k} \times \mathbf{j} &= -\mathbf{i} \\
\\
\mathbf{i} \times \mathbf{i} &= \mathbf{j} \times \mathbf{j} = \mathbf{k} \times \mathbf{k} = \vec{0}
\end{aligned}
$$

##### Perkalian Titik (Dot Product):
$$
\vec{v} \cdot \vec{u} = v_1 u_1 + v_2 u_2 + v_3 u_3
$$

Contoh:  
$$
\vec{v} =
\begin{pmatrix}
1 \\ 1 \\ 1
\end{pmatrix},
\quad
\vec{u} =
\begin{pmatrix}
1 \\ 2 \\ 1
\end{pmatrix}
\Rightarrow
\vec{v} \cdot \vec{u} = 1\cdot1 + 1\cdot2 + 1\cdot1 = 4
$$

##### Contoh Cross Product Lanjutan

Diketahui:
$$
\vec{u} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}, \quad
\vec{v} = \begin{pmatrix} 2 \\ 1 \\ 1 \end{pmatrix}
$$

Maka:
$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 1 & 1 \\
2 & 1 & 1
\end{vmatrix}
=
(1\cdot1 - 1\cdot1)\mathbf{i}
- (1\cdot1 - 1\cdot2)\mathbf{j}
+ (1\cdot1 - 1\cdot2)\mathbf{k}
$$

$$
= (0)\mathbf{i} - (-1)\mathbf{j} + (-1)\mathbf{k}
= \mathbf{j} - \mathbf{k}
$$

Atau dalam bentuk vektor kolom:
$$
\vec{u} \times \vec{v} = \begin{pmatrix} 0 \\ 1 \\ -1 \end{pmatrix}
$$

Catatan:
$$
\mathbf{i} \times \mathbf{j} \ne \mathbf{j} \times \mathbf{i}
$$


##### SOAL 1


##### 1. Luas dari jajaran genjang yang ditentukan oleh vektor **u = [1, 2]** dan **v = [2, 1]** adalah ‚Ä¶

**Jawaban:**

$$
\text{Luas} = |u \times v| = |(1)(1) - (2)(2)| = |1 - 4| = \boxed{3}
$$


##### 2. Luas dari jajaran genjang yang ditentukan oleh vektor **u = [2, 0, 0]** dan **v = [0, 0, 3]** adalah ‚Ä¶

**Jawaban:**

$$
u \times v = 
\begin{bmatrix}
0 \cdot 3 - 0 \cdot 0 \\
-(2 \cdot 3 - 0 \cdot 0) \\
2 \cdot 0 - 0 \cdot 0
\end{bmatrix}
=
\begin{bmatrix}
0 \\
-6 \\
0
\end{bmatrix}
$$

$$
|u \times v| = \sqrt{0^2 + (-6)^2 + 0^2} = \sqrt{36} = \boxed{6}
$$


##### 3. Luas segitiga dengan titik sudut **A(0,0,0)**, **B(1,3,-1)**, dan **C(2,1,1)** adalah ‚Ä¶

**Jawaban:**

$$
\vec{AB} = \langle 1, 3, -1 \rangle \\
\vec{AC} = \langle 2, 1, 1 \rangle
$$

$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
i & j & k \\
1 & 3 & -1 \\
2 & 1 & 1
\end{vmatrix}
=
i(3 \cdot 1 - (-1)\cdot1) - j(1\cdot1 - (-1)\cdot2) + k(1\cdot1 - 3\cdot2)
$$

$$
= i(4) - j(3) + k(-5) = \langle 4, -3, -5 \rangle
$$

$$
|\vec{AB} \times \vec{AC}| = \sqrt{4^2 + (-3)^2 + (-5)^2} = \sqrt{16 + 9 + 25} = \sqrt{50}
$$

$$
\text{Luas segitiga} = \frac{1}{2} \cdot \sqrt{50} = \boxed{\frac{\sqrt{50}}{2} \approx 3.54}
$$


##### 4. Luas segitiga dengan titik sudut **A(5,2,-1)**, **B(3,6,2)**, dan **C(1,0,4)** adalah ‚Ä¶

**Jawaban:**

$$
\vec{AB} = \langle -2, 4, 3 \rangle \\
\vec{AC} = \langle -4, -2, 5 \rangle
$$

$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
i & j & k \\
-2 & 4 & 3 \\
-4 & -2 & 5
\end{vmatrix}
=
i(4\cdot5 - 3\cdot(-2)) - j(-2\cdot5 - 3\cdot(-4)) + k(-2\cdot(-2) - 4\cdot(-4))
$$

$$
= i(20 + 6) - j(-10 + 12) + k(4 + 16) = \langle 26, -2, 20 \rangle
$$

$$
|\vec{AB} \times \vec{AC}| = \sqrt{26^2 + (-2)^2 + 20^2} = \sqrt{676 + 4 + 400} = \sqrt{1080}
$$

$$
\text{Luas segitiga} = \frac{1}{2} \cdot \sqrt{1080} = \boxed{\frac{\sqrt{1080}}{2} \approx 16.43}
$$
