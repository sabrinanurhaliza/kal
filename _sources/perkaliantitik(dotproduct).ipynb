# PERKALIAN TITIK (Dot Product)

##### Definisi

Perkalian titik adalah operasi aljabar sederhana yang membantu kita mengukur panjang vektor dan sudut yang dibentuk oleh sepasang vektor.

## Dot Product Basis 2D dan 3D

### Basis 2D
Terdapat dua vektor basis di ℝ²:

$$
\mathbf{b}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad 
\mathbf{b}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

Hitung dot product-nya:

$$
\mathbf{b}_1 \cdot \mathbf{b}_2 = (1 \cdot 0) + (0 \cdot 1) = 0
$$

Karena hasilnya 0, maka $\mathbf{b}_1$ dan $\mathbf{b}_2$ **ortogonal**.



### Basis 3D
Terdapat tiga vektor basis di ℝ³:

$$
\mathbf{b}_1 = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}, \quad
\mathbf{b}_2 = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}, \quad
\mathbf{b}_3 = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}
$$

#### Hitung semua pasangan dot product:

1. $\mathbf{b}_1 \cdot \mathbf{b}_2$
   $$
   = (1 \cdot 0) + (0 \cdot 1) + (0 \cdot 0) = 0
   $$

2. $\mathbf{b}_1 \cdot \mathbf{b}_3$
   $$
   = (1 \cdot 0) + (0 \cdot 0) + (0 \cdot 1) = 0
   $$

3. $\mathbf{b}_2 \cdot \mathbf{b}_3$
   $$
   = (0 \cdot 0) + (1 \cdot 0) + (0 \cdot 1) = 0
   $$

Semua pasangan dot product hasilnya 0, sehingga ketiga vektor **saling ortogonal**.



##### Kesimpulan
- Vektor basis standar di 2D dan 3D adalah **saling ortogonal**.
- Dot product dua vektor ortogonal selalu bernilai **nol (0)**.

# Vektor Satuan di $\mathbb{R}^n$

**Definisi**

Sebuah vektor $\mathbf{u}$ disebut sebagai **vektor satuan** jika norm (atau panjang) dari vektor tersebut adalah 1. Secara matematis:

$$\| \mathbf{u} \| = 1$$

Di ruang $\mathbb{R}^n$, untuk vektor $\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$, normnya adalah:

$$\| \mathbf{v} \| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$$

Untuk mendapatkan vektor satuan $\mathbf{u}$ yang searah dengan vektor tak-nol $\mathbf{v}$, kita normalisasi $\mathbf{v}$:

$$\mathbf{u} = \frac{\mathbf{v}}{\| \mathbf{v} \|}$$

**Contoh di $\mathbb{R}^2$**

Diberikan vektor $\mathbf{v} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}$.

1.  Hitung norm dari $\mathbf{v}$:
    $$\| \mathbf{v} \| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$$

2.  Vektor satuan $\mathbf{u}$ searah dengan $\mathbf{v}$ adalah:
    $$\mathbf{u} = \frac{1}{5} \begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 3/5 \\ 4/5 \end{bmatrix}$$

3.  Verifikasi bahwa $\mathbf{u}$ adalah vektor satuan:
    $$\| \mathbf{u} \| = \sqrt{\left(\frac{3}{5}\right)^2 + \left(\frac{4}{5}\right)^2} = \sqrt{\frac{9}{25} + \frac{16}{25}} = \sqrt{\frac{25}{25}} = 1$$

**Vektor Basis Standar di $\mathbb{R}^n$**

Vektor-vektor basis standar di $\mathbb{R}^n$ merupakan contoh vektor satuan yang saling ortogonal:

$$
\mathbf{e}_1 = \begin{bmatrix} 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}, \quad
\mathbf{e}_2 = \begin{bmatrix} 0 \\ 1 \\ \vdots \\ 0 \end{bmatrix}, \quad
\dots, \quad
\mathbf{e}_n = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 1 \end{bmatrix}
$$

Setiap vektor basis standar $\mathbf{e}_i$ memiliki norm 1:

$$\| \mathbf{e}_i \| = 1$$
