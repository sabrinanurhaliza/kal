```python
%matplotlib inline  
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np

titik = np.array([[0,0],[0.5,0.5],[0.5,1.5],[0,1],[0,0]])
koordinats = titik.transpose()
x = koordinats[0,:]
y = koordinats[1,:]

# Refleksi terhadap garis y = 2
x_LT2 = x
y_LT2 = 4 - y  # refleksi terhadap y = 2

# Membuat gambar dan sumbu sumbu
fig, ax = plt.subplots()

# Ploting titik titik. x dan y adalah titik asal, x_LT2 dan y_LT2 adalah bayangan (codomain)
ax.plot(x, y, 'ro', label='Titik Asli')
ax.plot(x_LT2, y_LT2, 'bo', label='Bayangan (Refleksi y = 2)')

# Menghubungkan titik dengan baris
ax.plot(x, y, 'r--')
ax.plot(x_LT2, y_LT2, 'b')

# Menambahkan garis refleksi y = 2
ax.axhline(y=2, color='g', ls='-', label='Garis y = 2')

# Sumbu koordinat
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")
ax.grid(True)
ax.axis([-2, 2, -1, 5])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = 2")
ax.legend()
```

```python
%matplotlib inline  
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np

# Titik asal
titik = np.array([[0,0],[0.5,0.5],[0.5,1.5],[0,1],[0,0]])
koordinats = titik.transpose()
x = koordinats[0,:]
y = koordinats[1,:]

# Refleksi terhadap garis y = x (menukar x dan y)
x_LT2 = y  # refleksi terhadap y = x (menukar x dan y)
y_LT2 = x

# Membuat gambar dan sumbu sumbu
fig, ax = plt.subplots()

# Ploting titik titik. x dan y adalah titik asal, x_LT2 dan y_LT2 adalah bayangan (codomain)
ax.plot(x, y, 'ro', label='Titik Asli')
ax.plot(x_LT2, y_LT2, 'bo', label='Bayangan (Refleksi y = x)')

# Menghubungkan titik dengan baris
ax.plot(x, y, 'r--')
ax.plot(x_LT2, y_LT2, 'b')

# Menambahkan garis refleksi y = x
x_garis = np.linspace(-2, 2, 100)
y_garis = x_garis
ax.plot(x_garis, y_garis, 'g-', label='Garis y = x')

# Sumbu koordinat
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")
ax.grid(True)
ax.axis([-2, 2, -1, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = x")
ax.legend()
```
