# MATRIK REFLECTION

Contoh transformasi refleksi menggunakan 5 matriks, lengkap dengan titik asal dan hasil transformasi untuk masing-masing jenis refleksi:

##### 1. Reflection about the x-axis

Matriks:
$\begin{bmatrix}  
    1 & 0\\
    0 & -1 
\end{bmatrix}$

Titik Asal → Titik Hasil:

(1, 4) → (1, -4)

(-3, 2) → (-3, -2)

##### 2. Reflection about the y-axis

Matriks:
$\begin{bmatrix}  
    -1 & 0\\
    0 & 1 
\end{bmatrix}$

Titik Asal → Titik Hasil:

(2, -5) → (-2, -5)

(-4, 1) → (4, 1)
