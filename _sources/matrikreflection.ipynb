```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Titik-titik baru (bentuk tetap serupa tapi angka berbeda)
titik = np.array([[0, 1], [1, 1.5], [1, 2.5], [0, 2], [0, 1]])
koordinats = titik.transpose()
x = koordinats[0, :]
y = koordinats[1, :]

# Refleksi terhadap garis y = 2
x_refleksi = x
y_refleksi = 4 - y  # y_refleksi = 2*2 - y

# Membuat plot
fig, ax = plt.subplots()

# Titik asli
ax.plot(x, y, 'ro', label='Titik Asli')
ax.plot(x, y, 'r--')

# Bayangan hasil refleksi
ax.plot(x_refleksi, y_refleksi, 'bo', label='Bayangan (Refleksi y = 2)')
ax.plot(x_refleksi, y_refleksi, 'b')

# Garis refleksi y = 2
ax.axhline(y=2, color='g', ls='-', label='Garis y = 2')

# Garis sumbu
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Tampilan
ax.grid(True)
ax.axis([-2, 2, -1, 5])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = 2")
ax.legend()

plt.show()
```

```python
# %matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np

# Titik asal baru (bentuk tetap tapi koordinat berbeda)
titik = np.array([[1, 0], [1.5, 0.5], [1.5, 1.5], [1, 1], [1, 0]])
koordinats = titik.transpose()
x = koordinats[0, :]
y = koordinats[1, :]

# Refleksi terhadap garis x = 2
x_refleksi = 4 - x  # karena x_refleksi = 2*2 - x
y_refleksi = y

# Membuat gambar dan sumbu
fig, ax = plt.subplots()

# Titik asli
ax.plot(x, y, 'ro', label='Titik Asli')
ax.plot(x, y, 'r--')

# Bayangan hasil refleksi
ax.plot(x_refleksi, y_refleksi, 'bo', label='Bayangan (Refleksi x = 2)')
ax.plot(x_refleksi, y_refleksi, 'b')

# Garis refleksi x = 2
ax.axvline(x=2, color='g', ls='-', label='Garis x = 2')

# Sumbu koordinat
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Tampilan
ax.grid(True)
ax.axis([-2, 6, -1, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis x = 2")
ax.legend()

plt.show()
```

##### y = x

```python
%matplotlib inline  
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np

titik = np.array([[0, 1], [0.5, 1.5], [0.5, 2.5], [0, 2], [0, 1]])
koordinats = titik.transpose()
x = koordinats[0,:]
y = koordinats[1,:]
```

```python
# %matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np

# Titik asal baru (ubah nilai tapi bentuk tetap sama)
titik = np.array([[0, 1], [0.5, 1.5], [0.5, 2.5], [0, 2], [0, 1]])
koordinats = titik.transpose()
x = koordinats[0, :]
y = koordinats[1, :]

# Refleksi terhadap garis y = x (dengan menukar x dan y)
x_refleksi = y
y_refleksi = x

# Membuat gambar dan sumbu
fig, ax = plt.subplots()

# Titik asli
ax.plot(x, y, 'ro', label='Titik Asli')
ax.plot(x, y, 'r--')

# Bayangan hasil refleksi
ax.plot(x_refleksi, y_refleksi, 'bo', label='Bayangan (Refleksi y = x)')
ax.plot(x_refleksi, y_refleksi, 'b')

# Garis y = x
x_garis = np.linspace(-2, 3, 100)
y_garis = x_garis
ax.plot(x_garis, y_garis, 'g-', label='Garis y = x')

# Sumbu koordinat
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Tampilan
ax.grid(True)
ax.axis([-2, 3, -1, 3])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = x")
ax.legend()

plt.show()
```
